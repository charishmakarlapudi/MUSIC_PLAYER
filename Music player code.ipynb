```python
## music player
import pygame
import tkinter as tk
from tkinter.filedialog import askdirectory
import os

music_player = tk.Tk()
music_player.title("My Music Player")
music_player.geometry("450x350")

directory = askdirectory()
os.chdir(directory)
song_list = os.listdir()

play_list = tk.Listbox(music_player, font="Helvetica 12 bold", bg='yellow', selectmode=tk.SINGLE)
for item in song_list:
    pos = 0
    play_list.insert(pos, item)
    pos += 1

pygame.init()
pygame.mixer.init()

def play_music():
    pygame.mixer.music.load(play_list.get(tk.ACTIVE))
    var.set(play_list.get(tk.ACTIVE))
    pygame.mixer.music.play()

def stop_music():
    pygame.mixer.music.stop()

def pause_music():
    pygame.mixer.music.pause()

def unpause_music():
    pygame.mixer.music.unpause()

button_play = tk.Button(music_player, width=5, height=3, font="Helvetica 12 bold", text="PLAY",
                        command=play_music, bg="green", fg="white")
button_stop = tk.Button(music_player, width=5, height=3, font="Helvetica 12 bold", text="STOP",
                        command=stop_music, bg="red", fg="white")
button_pause = tk.Button(music_player, width=5, height=3, font="Helvetica 12 bold", text="PAUSE",
                         command=pause_music, bg="blue", fg="white")
button_unpause = tk.Button(music_player, width=5, height=3, font="Helvetica 12 bold", text="UNPAUSE",
                           command=unpause_music, bg="orange", fg="white")

var = tk.StringVar()
song_title = tk.Label(music_player, font="Helvetica 12 bold", textvariable=var)
song_title.pack()

button_play.pack(fill="x")
button_stop.pack(fill="x")
button_pause.pack(fill="x")
button_unpause.pack(fill="x")
play_list.pack(fill="both", expand="yes")

music_player.mainloop()
```
